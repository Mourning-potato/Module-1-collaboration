```python
class Solution1:
    def sort012(self, arr):
        length = len(arr)
        for i in range(length):
            for ii in range(0, length - i - 1):
                dummy = 0
                if arr[ii] > arr[ii + 1]:
                    dummy = arr[ii]
                    arr[ii] = arr[ii + 1]
                    arr[ii + 1] = dummy

class Solution2:
    def binarysearch(self, arr, k):
            length = len(arr)
            for i in range(0, length):
                if arr[i] == k:
                    return i
            return -1
```
