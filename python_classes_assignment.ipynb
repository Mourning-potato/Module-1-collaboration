```python
# date: 11/15/2025
# author: Edward B. Bowen Jr.
# purpose: to make a class for a vehicle, and a subclass for automobile
class Vehicle:
    vehicle_type = None

class Automobile(Vehicle):
    Vehicle.vehicle_type = "car"
    year = None
    make = None
    model = None
    doors = None
    roof = None
    def __init__(self, year_arg, make_arg, model_arg, door_arg, roof_arg):
        print(self.vehicle_type)
        self.year = year_arg
        self.make = make_arg
        self.model = model_arg
        self.doors = door_arg
        self.roof = roof_arg

        self.output()

    def output(self):
        print("Vehicle type: " + self.vehicle_type)
        print("Year: " + str(self.year))
        print("Make: " + self.make)
        print("Model: " + self.model)
        print("Number of doors: " + str(self.doors))
        print("Type of roof: " + self.roof)

year = input("Input the year the car was made ")
make = input("Input the maker of the car ")
model = input("Input the model of the car ")
doors = input("Input the number of doors the car has ")
roof = input("Input either sun roof, or solid roof, depending on the roof type of the car ")

my_car = Automobile(year, make, model, doors, roof)
```
